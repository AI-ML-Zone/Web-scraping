### Q9: Best All-round performers
#### Use bat avg > 20 and bowling avg in between 20 to 30 as criteria

##### To judge allrounder, bowling should be least and batting average should be highest

```python
df3['bat_avg'] = df3['bat_avg'].astype('int64')
df3['bw_avg'] = df3['bw_avg'].astype('int64')
```

```python
df3_new = df3[['Player', 'bat_avg','bw_avg']].copy()
df3_new
```

```python
df3_new['bw_avg'] = df3['bw_avg'].apply(lambda x:'yes' if (x>20 and x<30) else 'No')
df3_new['bat_avg'] = df3['bat_avg'].apply(lambda x:'yes' if (x>20) else 'No')
```

### Q19 . Plot a pie chart for displaying the England run percentage

```python
df3_eng = df3.loc[0:17,['Player','runs_scored']]
df3_eng
```

```python
df3_eng['var2'] = pd.Series([round(val, 2) for val in df3_eng['runs_scored']], index = df3_eng.index)
df3_eng['var3'] = pd.Series(['{0:.1f}%'.format(val * 100) for val in df3_eng['var2']], index = df3_eng.index)
```

```python
df3_eng['var3']
```

```python
h = df3_eng['var3'].str.extract('(\d+)').astype(float)
h
```

```python
import matplotlib.pyplot as plt
plt.pie(df3_eng['runs_scored'],labels=df3_eng['Player'],autopct='%.1f%%',radius=4.5)
plt.show()
```
